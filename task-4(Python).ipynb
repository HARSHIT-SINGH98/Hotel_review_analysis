```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('Hotel_Reviews.csv')
```

```python
df
```

```python
df.isnull()
```

```python
# to drop the column which ha
df.drop(["Hotel_Address"], axis = 1)

# Set the number of decimal places you want to keep 
num_decimal_places = 2

# Round the values in the desired column to the specified number of decimal places
df['lat'] = df['lat'].round(num_decimal_places)
df['lng'] = df['lng'].round(num_decimal_places)
df
```

```python
df.info()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
sns.heatmap(df.corr(), annot = True)
```

```python
sorted_hotels= df.sort_values("Reviewer_Score")
```

```python
sorted_hotels
```

```python
# Top 10 worst hotels 
worst_hotels = sorted_hotels.head(10)
worst_hotels
```

```python
best_hotels = sorted_hotels.tail(10)
best_hotels
```

# Bar Plot B/w TOP 10 wrost and best hotels according to 
# Reviewer Score

```python

fig, ax = plt.subplots(figsize=(10, 5))
ax.bar(worst_hotels['Hotel_Name'], worst_hotels['Reviewer_Score'], color='red')
ax.bar(best_hotels['Hotel_Name'], best_hotels['Reviewer_Score'], color='green')
ax.set_title('Worst Top 10 Hotels and Top Best Hotel by Reviewer Score')
ax.set_xlabel('Hotel Name')
ax.set_ylabel('Reviewer Score')

# Rotate the x-axis labels for readability
plt.xticks(rotation=90)

# Show the plot
plt.show()
```
